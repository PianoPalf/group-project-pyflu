```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
```

```python
#create path to csv file:
hygiene_path=path('Resources\Hygiene data.csv') 
```

```python
hygiene_df=pd.read_csv(hygiene_path)
hygiene_df.head()
```

```python
# Rename the columns
new_columns = {
    '2017 [YR2017]': '2017',
    '2018 [YR2018]': '2018',
    '2019 [YR2019]': '2019',
    '2020 [YR2020]': '2020',
    '2021 [YR2021]': '2021',
    '2022 [YR2022]': '2022'
}
hygiene_df= hygiene_df.rename(columns=new_columns)
```

```python
hygiene_df.head()
```

```python
# Select specific columns
columns_to_keep = ['Country Code', 'Country Name', '2017', '2018', '2019', '2020', '2021', '2022']
hygiene_df = hygiene_df[columns_to_keep]

# Set 'Country Name' as the index
hygiene_df = hygiene_df.set_index(keys=['Country Code','Country Name',])

# Stack the data
hygiene_df = hygiene_df.stack()

# Convert the stacked data to a DataFrame
hygiene_df = pd.DataFrame(hygiene_df)

# Reset the index to get a default numeric index
hygiene_df = hygiene_df.reset_index()

```

```python
# Rename the columns

hygiene_df.columns = ['Country Code', 'Country Name',  'Year', 'People With Handwashing Facilities']
```

```python
hygiene_df.head(10)
```

```python
#Drop missing value
hygiene_df.replace('..', np.nan, inplace=True)
new_hygiene_df =hygiene_df.dropna()
new_hygiene_df.head(15)
```

```python
# Write the new data to a CSV file namely: hygiene cleaned
    #save as hygiene cleaned file
newfile_path= path('Resources\hygiene cleaned.csv')
new_hygiene_df.to_csv(newfile_path, index=False)
```
