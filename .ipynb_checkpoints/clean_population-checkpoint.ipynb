```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# Read the CSV file into 'pop' DataFrame (to use to calculate per capita rate)
populations = Path("Resources/populations_countries.csv")

populations_df = pd.read_csv(populations, usecols = ['Country Name', 
                                                     'Country Code', 
                                                     '2017 [YR2017]', 
                                                     '2018 [YR2018]',
                                                     '2019 [YR2019]', 
                                                     '2020 [YR2020]',
                                                     '2021 [YR2021]',
                                                     '2022 [YR2022]'])

# Rename Columns
populations_df.rename(columns={'2017 [YR2017]': '2017', \
                              '2018 [YR2018]': '2018', \
                              '2019 [YR2019]': '2019', \
                              '2020 [YR2020]': '2020',
                              '2021 [YR2021]': '2021',
                              '2022 [YR2022]': '2022'}, \
                              inplace=True)

# Change multiple Year Columns to one Column, to match 'influenza_df'
populations_one_column_df = populations_df.melt(id_vars=["Country Name", 'Country Code'], 
                                        var_name="Year", 
                                        value_name="Population")

populations_one_column_df
```

```python
# Save the populations_one_column_df DataFrame to a CSV file
populations_one_column_df.to_csv('Resources/populations_one_column.csv', index=False)
```
