```python
keep_colunms= ['COUNTRY_CODE', 'COUNTRY_AREA_TERRITORY', 'ISO_YEAR','INF_ALL']
influenza_df= influenza_df[keep_colunms]
```

```python
influenza_df
```

```python
# Rename some columns for merging later
new_columns = {
    'COUNTRY_CODE': 'Country Code',
    'COUNTRY_AREA_TERRITORY': 'Country Name',
    'ISO_YEAR': 'Year',
    'INF_ALL': 'Influenza Cases'
}

influenza_df = influenza_df.rename(columns=new_columns)
influenza_df
```

```python
influenza_df = influenza_df.dropna()
influenza_df= influenza_df.reset_index(drop=True)
influenza_df.head(10)
```

```python
influenza_df = influenza_df[influenza_df['Year'].isin([2017, 2018, 2019, 2020, 2021, 2022])]
influenza_df= influenza_df.reset_index(drop=True)
influenza_df.head(10)
```

```python
# Write the new data to a CSV file
output_file_path=path('Resources\influenza cleaned.csv')
influenza_df.to_csv(output_file_path, index=False)
```
