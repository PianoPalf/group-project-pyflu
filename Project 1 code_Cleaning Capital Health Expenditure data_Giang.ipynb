```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
```

```python

#create path to csv file:
expenditure_path= path('Resources\Capital health expenditure_data.csv')
```

```python
expenditure_df= pd.read_csv(expenditure_path)
expenditure_df.head()
```

```python
# Rename the columns
new_columns = {
    '2017 [YR2017]': '2017',
    '2018 [YR2018]': '2018',
    '2019 [YR2019]': '2019',
    '2020 [YR2020]': '2020',
    '2021 [YR2021]': '2021',
    '2022 [YR2022]': '2022'
}
expenditure_df = expenditure_df.rename(columns=new_columns)


```

```python
# Rename the columns
new_columns = {
    '2017 [YR2017]': '2017',
    '2018 [YR2018]': '2018',
    '2019 [YR2019]': '2019',
    '2020 [YR2020]': '2020',
    '2021 [YR2021]': '2021',
    '2022 [YR2022]': '2022'
}
expenditure_df.rename(columns=new_columns)
# Check the modified DataFrame
expenditure_df.head()
```

```python
# Select specific columns
columns_to_keep = ['Country Code', 'Country Name', '2017', '2018', '2019', '2020', '2021', '2022']
expenditure_df = expenditure_df[columns_to_keep]

# Set 'Country Name' as the index
expenditure_df = expenditure_df.set_index(keys=['Country Code','Country Name',])

# Stack the data
expenditure_df = expenditure_df.stack()

# Convert the stacked data to a DataFrame
expenditure_df = pd.DataFrame(expenditure_df)

# Reset the index to get a default numeric index
expenditure_df = expenditure_df.reset_index()

expenditure_df 
```

```python
# Rename the columns

expenditure_df.columns = ['Country Code', 'Country Name',  'Year', 'Capital Health Expenditure']
```

```python
expenditure_df.head(10)
```

```python
#Drop missing value
expenditure_df.replace('..', np.nan, inplace=True)
new_expenditure_df =expenditure_df.dropna()
new_expenditure_df.head(15)
```

```python
# Write the new data to a CSV file namely: hygiene cleaned

newfile_path= path('Resources\Capital Health Expenditure cleaned.csv')
new_expenditure_df.to_csv(newfile_path, index=False)
```
